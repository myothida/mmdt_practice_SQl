### Tutorial 10: Data Analysis of  Wide World Importers: Sales Report


#### **Problem Set**

### Analytical Questions for Comprehensive Report

#### 1. Order Overview
1. **Total Orders**: How many total orders are in the dataset?
2. **Order Trends Over Time**:
   - How many orders were placed each month/year?
   - Are there specific months or years with significantly higher or lower order volumes?
3. **Order Completion**:
   - What percentage of orders have been completed (e.g., `PickingCompletedWhen` is not null)?
   - What is the average time between `OrderDate` and `PickingCompletedWhen`?

---

#### 2. Customer Analysis
4. **Customer Distribution**:
   - How many unique customers (`CustomerID`) are there?
   - Which customers have placed the most orders? (Top 10 customers by order count.)
5. **Customer Loyalty**:
   - How many orders do most customers place? (Distribution of orders per customer.)
   - Are there customers who place orders frequently (e.g., monthly or quarterly)?
6. **Customer Backorders**:
   - Which customers have the most backordered orders (`BackorderOrderID` not null)?
   - What percentage of orders per customer are backordered?

---

#### 3. Salesperson Performance
7. **Salesperson Activity**:
   - How many unique salespersons (`SalespersonPersonID`) are there?
   - Which salespersons have processed the most orders? (Top 10 salespersons by order count.)
8. **Salesperson Efficiency**:
   - What is the average time between `OrderDate` and `PickingCompletedWhen` for each salesperson?
   - Are there salespersons with a high number of backordered orders?
9. **Salesperson Trends**:
   - Are there salespersons whose order volumes have increased or decreased over time?

---

#### 4. Picking and Delivery Analysis
10. **Picking Efficiency**:
    - What is the average time between `OrderDate` and `PickingCompletedWhen`?
    - Are there specific months or years where picking times were longer?
11. **Delivery Instructions**:
    - How many orders have `DeliveryInstructions` filled out?
    - What are the most common delivery instructions (e.g., "Fragile," "Leave at door")?
12. **Delivery Delays**:
    - What is the average time between `ExpectedDeliveryDate` and `PickingCompletedWhen`?
    - How many orders were delivered later than the `ExpectedDeliveryDate`?

---

#### 5. Backorder Analysis
13. **Backorder Frequency**:
    - How many orders are backordered (`BackorderOrderID` not null)?
    - What percentage of total orders are backordered?
14. **Backorder Trends**:
    - Are there specific months or years with higher backorder rates?
    - Are there specific customers or products associated with frequent backorders?
15. **Backorder Resolution**:
    - What is the average time between `OrderDate` and `PickingCompletedWhen` for backordered orders?
    - How many backordered orders are still pending completion?

---

#### 6. Undersupply Analysis
16. **Undersupply Frequency**:
    - How many orders have `IsUndersupplyBackordered` set to True?
    - What percentage of total orders are affected by undersupply?
17. **Undersupply Impact**:
    - What is the average time between `OrderDate` and `PickingCompletedWhen` for undersupplied orders?
    - Are there specific customers or products associated with frequent undersupply issues?

---

#### 7. Comments and Feedback Analysis
18. **Customer Comments**:
    - How many orders have `Comments` filled out by customers?
    - What are the most common words or phrases in `Comments`? (Use text analysis.)
19. **Internal Comments**:
    - How many orders have `InternalComments` filled out?
    - What are the most common words or phrases in `InternalComments`? (Use text analysis.)
20. **Feedback Trends**:
    - Are there specific months or years with higher volumes of customer or internal comments?

---

#### 8. Editing and Updates Analysis
21. **Order Edits**:
    - How many orders have been edited (`LastEditedBy` not null)?
    - Who (`LastEditedBy`) is making the most edits to orders?
22. **Edit Trends**:
    - Are there specific months or years with higher edit activity?
    - What is the average time between `OrderDate` and `LastEditedWhen`?

---

#### 9. Temporal Analysis
23. **Seasonal Trends**:
    - Are there specific days of the week, months, or seasons with higher order volumes?
    - Are there specific days of the week, months, or seasons with longer delivery times?
24. **Year-over-Year Growth**:
    - How has the total number of orders changed year-over-year?
    - How has the average delivery time changed year-over-year?

---

#### 10. Data Quality and Completeness
25. **Missing Data**:
    - How many orders have missing values in critical columns like `CustomerID`, `OrderDate`, or `ExpectedDeliveryDate`?
    - Are there patterns in missing data (e.g., specific months or customers)?
26. **Data Consistency**:
    - Are there any orders where `ExpectedDeliveryDate` is earlier than `OrderDate`?
    - Are there any orders where `PickingCompletedWhen` is earlier than `OrderDate`?

---

#### 11. Advanced Analysis
27. **Customer Segmentation**:
    - Can customers be segmented based on order frequency, order value, or backorder rates?
28. **Salesperson-Customer Relationships**:
    - Are there specific salespersons who handle orders for specific customers?
29. **Predictive Analysis**:
    - Can we predict which orders are likely to be backordered based on historical data?

---

#### Example Report Structure
1. **Executive Summary**: Key findings and insights.
2. **Order Overview**: Trends, volumes, and completion rates.
3. **Customer Insights**: Top customers, loyalty, and backorder patterns.
4. **Salesperson Performance**: Top performers and efficiency metrics.
5. **Delivery and Picking Analysis**: Efficiency and delays.
6. **Backorder and Undersupply Analysis**: Frequency, trends, and impact.
7. **Comments and Feedback**: Customer and internal feedback trends.
8. **Temporal Analysis**: Seasonal trends and year-over-year growth.
9. **Data Quality**: Missing data and consistency checks.
10. **Recommendations**: Actionable insights for improving operations.
